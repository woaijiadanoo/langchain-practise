```python
# LLM的标准事件 - with_structured_output
# 影响LLM的输出：以结构化的数据来输出

from typing import Optional
from pydantic import BaseModel, Field
from langchain_deepseek import ChatDeepSeek
import os

# 创建DeepSeek 聊天模型实例
llm = ChatDeepSeek(
    temperature=0,  # temperature=0表示输出更加确定性，不会随机性太强
    model_name="deepseek-chat",  # 模型名称
    api_key=os.getenv("DEEPSEEK_API_KEY")  # 从环境变量中获取API密钥
)

# Pydantic数据模型
class Joke(BaseModel):
    """Joke to tell user."""
    setup: str = Field(description="The setup of the joke")
    punchline: str = Field(description="The punchline to the joke")
    rating: Optional[int] = Field(
        default=None, description="How funny the joke is, from 1 to 10"
    )

# 将LLM配置为输出结构化数据
structured_llm = llm.with_structured_output(Joke)

# 调用模型生成结构化笑话
result = structured_llm.invoke("给我讲一个关于程序员的笑话")
print(result)
```

```python
# LLM的标准事件 - with_structured_output
# 影响LLM的输出：以结构化的数据来输出
from typing_extensions import Annotated, TypedDict
from typing import Optional, Union
from pydantic import BaseModel, Field
from langchain_deepseek import ChatDeepSeek
import os

# 创建DeepSeek 聊天模型实例
llm = ChatDeepSeek(
    temperature=0,  # temperature=0表示输出更加确定性，不会随机性太强
    model_name="deepseek-chat",  # 模型名称
    api_key=os.getenv("DEEPSEEK_API_KEY")  # 从环境变量中获取API密钥
)

# 定义Joke类（使用TypedDict）
class Joke(TypedDict):
    """Joke to tell user."""
    setup: Annotated[str, "The setup of the joke"]
    punchline: Annotated[str, "The punchline of the joke"]
    rating: Annotated[Optional[int], "How funny the joke is, from 1 to 10"]

# 将LLM配置为输出结构化数据
structured_llm = llm.with_structured_output(Joke)

# 使用流式输出生成笑话
for chunk in structured_llm.stream("给我讲一个关于程序员的笑话"):
    print(chunk)
```

```python
from langchain_deepseek import ChatDeepSeek
import os

# 创建DeepSeek 聊天模型实例
llm = ChatDeepSeek(
    temperature=0,  # temperature=0表示输出更加确定性，不会随机性太强
    model_name="deepseek-chat",  # 模型名称
    api_key=os.getenv("DEEPSEEK_API_KEY")  # 从环境变量中获取API密钥
)

question = "Introduction the langchain with one sentence."

result = llm.invoke(question)
## 优先使用 usage_metadata 获取Token数据
result.usage_metadata
## 每一家的模型都可以定义不同的Token获取字段
print(f'Deepseek : {result.response_metadata["token_usage"]}')
```

```python
from langchain_deepseek import ChatDeepSeek
import os

# 创建DeepSeek 聊天模型实例
llm = ChatDeepSeek(
    temperature=0,  # temperature=0表示输出更加确定性，不会随机性太强
    model_name="deepseek-chat",  # 模型名称
    api_key=os.getenv("DEEPSEEK_API_KEY")  # 从环境变量中获取API密钥
)

question = "Introduction the langchain with one sentence."

# 初始化聚合变量
aggregate = None

# 使用流式输出处理问题
for chunk in llm.stream(question, stream_usage=True):
    print(chunk)
    aggregate = chunk if aggregate is None else aggregate + chunk

# 打印最终聚合结果
print("\n聚合结果：")
print(aggregate)
```

```python
# 速率限制需要langchain-core>=0.2.24
# langchain中的InMemoryRateLimiter只能限制单位时间内的请求数量
# 无法处理根据请求大小来限制的情况
from langchain_core.rate_limiters import InMemoryRateLimiter
from langchain_anthropic import ChatAnthropic
import time
from langchain_deepseek import ChatDeepSeek
import os

# 创建内存速率限制器
rate_limiter = InMemoryRateLimiter(
    requests_per_second=1,  # 每1秒请求一次
    check_every_n_seconds=0.1,  # 每100毫秒检查一次是否允许
    max_bucket_size=10  # 控制最大突发大小
)

# 创建DeepSeek 聊天模型实例
llm = ChatAnthropic(
    model="deepseek-chat",  # 模型名称
    api_key=os.getenv("DEEPSEEK_API_KEY"),  # 从环境变量中获取API密钥
    rate_limiter=rate_limiter  # 请求速率限制
)

question = "Introduction the langchain with two words."

# 使用计时器来计算每次请求的时间间隔
for _ in range(5):
    tic = time.time()
    llm.invoke(question)
    toc = time.time()
    print(toc - tic)
```

```python
from pydantic import BaseModel, Field
from langchain_deepseek import ChatDeepSeek
# 导入环境变量加载工具
from dotenv import load_dotenv
import os

# 加载.env文件中的环境变量
load_dotenv()

# 加法工具
class add(BaseModel):
    """Add two integers."""
    a: int = Field(..., description="First integer")
    b: int = Field(..., description="Second integer")

# 乘法工具
class multiply(BaseModel):
    """Multiply two integers."""
    a: int = Field(..., description="First integer")
    b: int = Field(..., description="Second integer")

# 创建DeepSeek 聊天模型实例
llm = ChatDeepSeek(
    temperature=0,  # temperature=0表示输出更加确定性，不会随机性太强
    model_name="deepseek-chat",  # 模型名称
    api_key=os.getenv("DEEPSEEK_API_KEY")  # 从环境变量中获取API密钥
)

# 定义工具列表
tools = [add, multiply]

# 将工具绑定到语言模型
llm_with_tools = llm.bind_tools(tools)

# 定义查询并调用工具
query = "3乘以12是多少?"
result = llm_with_tools.invoke(query)

# 获取工具调用结果
print(result.tool_calls)
```
